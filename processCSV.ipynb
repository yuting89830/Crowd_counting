```python
import pandas as pd
from datetime import timedelta

# read origin data
df = pd.read_csv('/content/drive/MyDrive/colab用檔案/grafana_data_export.csv',header=1,parse_dates=True)
df
```

```python
# create dataframe
processed_df = pd.DataFrame(columns=['Time', 'A metric'])

# process origin data
for i in range(len(df)):
    current_row = df.iloc[i]
    processed_df = processed_df.append(current_row, ignore_index=True)

    if i < len(df) - 1:
        next_row = df.iloc[i + 1]
        current_time = pd.to_datetime(current_row['Time'])
        next_time = pd.to_datetime(next_row['Time'])
        time_diff = next_time - current_time
        missing_intervals = int(time_diff.total_seconds() / 60) - 1

        if missing_intervals > 0:
            for j in range(1, missing_intervals + 1):
                new_time = current_time + timedelta(minutes=j)
                avg_value = (current_row['A metric'] + next_row['A metric']) / 2
                processed_df = processed_df.append({'Time': new_time, 'A metric': avg_value}, ignore_index=True)

# save into process.csv
processed_df.to_csv('/content/drive/MyDrive/colab用檔案/grafana_data_process.csv', index=False)
```
